```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
# 1. Isolate Numerical Columns
numerical_cols = df.select_dtypes(include=['number']).columns

# 2. Fill Missing Numerical Values with Median
df[numerical_cols] = df[numerical_cols].fillna(df[numerical_cols].median())

# 3. Fill Missing Categorical Values
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    most_frequent = df[col].mode()[0]
    df[col].fillna(most_frequent, inplace=True)

# Data Visualization
# Let's say we are plotting distribution of 'contract_amount'
plt.figure(figsize=(10, 5))
plt.hist(df['contract_amount'], bins=50, color='blue', alpha=0.7, label='Contract Amount')
plt.title('Distribution of Contract Amount')
plt.xlabel('Contract Amount')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Calculate the 99th percentile of the contract_amount column
quantile_99 = df['contract_amount'].quantile(0.99)

# Filter out rows with contract_amount above the 99th percentile
df = df[df['contract_amount'] <= quantile_99]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Assuming df is your DataFrame
# df = pd.read_csv('your_dataset.csv')

# Disable scientific notation
pd.set_option('display.float_format', '{:.2f}'.format)

# Histogram
plt.figure(figsize=(10, 5))
sns.histplot(df['contract_amount'], bins=5000, kde=False)
plt.title('Histogram of Contract Amount')
plt.xlabel('Contract Amount')
plt.ylabel('Frequency')
plt.show()

# Boxplot to identify outliers
plt.figure(figsize=(10, 5))
sns.boxplot(x=df['contract_amount'])
plt.title('Boxplot of Contract Amount')
plt.xlabel('Contract Amount')
plt.show()

# Kernel Density Estimation (KDE) Plot
plt.figure(figsize=(10, 5))
sns.kdeplot(df['contract_amount'], shade=True)
plt.title('KDE Plot of Contract Amount')
plt.xlabel('Contract Amount')
plt.ylabel('Density')
plt.show()
```

```python
# Filter the DataFrame to only include rows where annual_amount is below 300000
df_filtered_annual = df[df['annual_amount'] < 300000]

# Histogram
plt.figure(figsize=(10, 5))
sns.histplot(df_filtered_annual['annual_amount'], bins=50, kde=False)
plt.title('Histogram of Annual Amount (< 300,000)')
plt.xlabel('Annual Amount')
plt.ylabel('Frequency')
plt.show()

# Boxplot to identify outliers
plt.figure(figsize=(10, 5))
sns.boxplot(x=df_filtered_annual['annual_amount'])
plt.title('Boxplot of Annual Amount (< 300,000)')
plt.xlabel('Annual Amount')
plt.show()

# KDE Plot
plt.figure(figsize=(10, 5))
sns.kdeplot(df_filtered_annual['annual_amount'], shade=True)
plt.title('KDE Plot of Annual Amount (< 300,000)')
plt.xlabel('Annual Amount')
plt.ylabel('Density')
plt.show()
```

```python
import numpy as np

# Calculate mean, median, and mode
mean_value = df['annual_amount'].mean()
median_value = df['annual_amount'].median()
mode_value = df['annual_amount'].mode()[0]  # Taking the first mode if multiple exist

# KDE Plot with annotations
plt.figure(figsize=(10, 5))
sns.kdeplot(df_filtered_annual['annual_amount'], shade=True)
plt.axvline(mean_value, color='r', linestyle='--', label=f'Mean: {mean_value:.2f}')
plt.axvline(median_value, color='g', linestyle='-', label=f'Median: {median_value:.2f}')
plt.axvline(mode_value, color='b', linestyle='-.', label=f'Mode: {mode_value:.2f}')
plt.title('KDE Plot of Annual Amount (< 300,000)')
plt.xlabel('Annual Amount')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you've read your DataFrame into 'df'

# 1. Annual Amount vs. Area
sns.scatterplot(data=df, x="area_name_en", y="annual_amount")
plt.title("Annual Amount by Area")
plt.xticks(rotation=90)
plt.show()

# 2. Contracts by Tenure
df['contract_start_date'] = pd.to_datetime(df['contract_start_date'])
df['contract_end_date'] = pd.to_datetime(df['contract_end_date'])
df['contract_duration'] = (df['contract_end_date'] - df['contract_start_date']).dt.days
sns.lineplot(data=df, x="contract_start_date", y="contract_duration")
plt.title("Contract Durations Over Time")
plt.show()

# 3. Distribution of Property Types
sns.countplot(data=df, x="ejari_property_type_en")
plt.title("Distribution of Property Types")
plt.xticks(rotation=90)
plt.show()

# 4. Correlation Heatmap
correlation_matrix = df[['annual_amount', 'no_of_prop', 'line_number', 'is_free_hold']].corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Heatmap")
plt.show()

# 5. Property Usage
df['property_usage_en'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Property Usage")
plt.show()
```

```python
# Convert the dates to datetime objects and create a 'contract_duration' column
df['contract_start_date'] = pd.to_datetime(df['contract_start_date'])
df['contract_end_date'] = pd.to_datetime(df['contract_end_date'])
df['contract_duration'] = (df['contract_end_date'] - df['contract_start_date']).dt.days

# 1. Group by Season
df['start_month'] = df['contract_start_date'].dt.month
seasons = {12: 'Winter', 1: 'Winter', 2: 'Winter',
           3: 'Spring', 4: 'Spring', 5: 'Spring',
           6: 'Summer', 7: 'Summer', 8: 'Summer',
           9: 'Autumn', 10: 'Autumn', 11: 'Autumn'}
df['start_season'] = df['start_month'].map(seasons)
sns.boxplot(data=df, x='start_season', y='contract_duration')
plt.title('Contract Duration by Season')
plt.show()

# 2. Group by Area
sns.boxplot(data=df, x='area_name_en', y='contract_duration')
plt.title('Contract Duration by Area')
plt.xticks(rotation=90)
plt.show()

# 3. Group by Property Type
sns.boxplot(data=df, x='ejari_property_type_en', y='contract_duration')
plt.title('Contract Duration by Property Type')
plt.xticks(rotation=90)
plt.show()

# 4. Relationship with Annual Amount
sns.scatterplot(data=df, x='contract_duration', y='annual_amount')
plt.title('Contract Duration vs. Annual Amount')
plt.show()
```
